```python
def batch(start, end):
    print(start, end)
    for k in range(start,end):
        i = data_files[k]
        df = pd.read_csv(i)
        print('Processing ' + str(i) + 'dataframe')

        for j in range(len(df)):
            
            if(df.iloc[j]['out_and_tx_malicious'] == 1):
                df_malicious_only =  df_malicious_only.append(df.iloc[j],ignore_index=True)
                
        print(str(i) + 'th datafile done')

    print(df_malicious_only)

```

```python
import threading
import os

global df_malicious_only
df_malicious_only = pd.DataFrame()

t1 = threading.Thread(target=batch, args=(0,1,))
t2 = threading.Thread(target=batch, args=(1,2,))
# starting thread 1
t1.start()
# starting thread 2
t2.start()

# wait until thread 1 is completely executed
t1.join()
# wait until thread 2 is completely executed
t2.join()

df_malicious_only
 
```
