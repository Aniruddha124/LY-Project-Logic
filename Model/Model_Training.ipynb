```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

def normalise(df):
    scaled_df = df.copy()
    scaled_df['in_degree'] = scaler.fit_transform(scaled_df['in_degree'].values.reshape(-1,1))
    scaled_df['out_degree']= scaler.fit_transform(scaled_df['out_degree'].values.reshape(-1,1))
    scaled_df['unique_in_degree']= scaler.fit_transform(scaled_df['out_degree'].values.reshape(-1,1))
    scaled_df['unique_out_degree']= scaler.fit_transform(scaled_df['out_degree'].values.reshape(-1,1))

    scaled_df['in_transaction_frequency']= scaler.fit_transform(scaled_df['in_transaction_frequency'].values.reshape(-1,1))
    scaled_df['out_transaction_frequency']= scaler.fit_transform(scaled_df['out_transaction_frequency'].values.reshape(-1,1))
    scaled_df['in_amount_frequency']= scaler.fit_transform(scaled_df['in_amount_frequency'].values.reshape(-1,1))
    scaled_df['out_amount_frequency']= scaler.fit_transform(scaled_df['out_amount_frequency'].values.reshape(-1,1))

    scaled_df['in_btc'] = scaler.fit_transform(scaled_df['in_btc'].values.reshape(-1,1))
    scaled_df['out_btc']= scaler.fit_transform(scaled_df['out_btc'].values.reshape(-1,1))
    scaled_df['balance']= scaler.fit_transform(scaled_df['balance'].values.reshape(-1,1))
    scaled_df['mean_in_btc']= scaler.fit_transform(scaled_df['mean_in_btc'].values.reshape(-1,1))
    scaled_df['mean_out_btc']= scaler.fit_transform(scaled_df['mean_out_btc'].values.reshape(-1,1))
    scaled_df['max_in_btc']= scaler.fit_transform(scaled_df['max_in_btc'].values.reshape(-1,1))
    scaled_df['min_in_btc']= scaler.fit_transform(scaled_df['min_in_btc'].values.reshape(-1,1))
    scaled_df['max_out_btc']= scaler.fit_transform(scaled_df['max_out_btc'].values.reshape(-1,1))
    scaled_df['min_out_btc']= scaler.fit_transform(scaled_df['min_out_btc'].values.reshape(-1,1))
    scaled_df['in_standard_deviation']= scaler.fit_transform(scaled_df['in_standard_deviation'].values.reshape(-1,1))
    scaled_df['out_standard_deviation']= scaler.fit_transform(scaled_df['out_standard_deviation'].values.reshape(-1,1))
    
    
    return scaled_df
```

```python
def log_normalise(df):
    scaled_df = df.copy()
    scaled_df['in_degree'] = np.log1p(scaled_df['in_degree'].values.reshape(-1,1))
    scaled_df['out_degree']= np.log1p(scaled_df['out_degree'].values.reshape(-1,1))
    scaled_df['unique_in_degree']= np.log1p(scaled_df['unique_in_degree'].values.reshape(-1,1))
    scaled_df['unique_out_degree']= np.log1p(scaled_df['unique_out_degree'].values.reshape(-1,1))

    scaled_df['in_transaction_frequency']= np.log1p(scaled_df['in_transaction_frequency'].values.reshape(-1,1))
    scaled_df['out_transaction_frequency']= np.log1p(scaled_df['out_transaction_frequency'].values.reshape(-1,1))
    scaled_df['in_amount_frequency']= np.log1p(scaled_df['in_amount_frequency'].values.reshape(-1,1))
    scaled_df['out_amount_frequency']= np.log1p(scaled_df['out_amount_frequency'].values.reshape(-1,1))

    scaled_df['in_btc'] = np.log1p(scaled_df['in_btc'].values.reshape(-1,1))
    scaled_df['out_btc']= np.log1p(scaled_df['out_btc'].values.reshape(-1,1))
    scaled_df['balance']= np.log1p(scaled_df['balance'].values.reshape(-1,1))
    scaled_df['mean_in_btc']= np.log1p(scaled_df['mean_in_btc'].values.reshape(-1,1))
    scaled_df['mean_out_btc']= np.log1p(scaled_df['mean_out_btc'].values.reshape(-1,1))
    scaled_df['max_in_btc']= np.log1p(scaled_df['max_in_btc'].values.reshape(-1,1))
    scaled_df['min_in_btc']= np.log1p(scaled_df['min_in_btc'].values.reshape(-1,1))
    scaled_df['max_out_btc']= np.log1p(scaled_df['max_out_btc'].values.reshape(-1,1))
    scaled_df['min_out_btc']= np.log1p(scaled_df['min_out_btc'].values.reshape(-1,1))
    scaled_df['in_standard_deviation']= np.log1p(scaled_df['in_standard_deviation'].values.reshape(-1,1))
    scaled_df['out_standard_deviation']= np.log1p(scaled_df['out_standard_deviation'].values.reshape(-1,1))
    
    return scaled_df
```
